```python
#Question 13
#drop the date and light column
energy = energy.drop(columns = ['date', 'lights'])
energy
```

```python
#normalising the dataset
scaler = MinMaxScaler()
normalised_data = pd.DataFrame(scaler.fit_transform(energy), columns= energy.columns)
features_data =normalised_data.drop(columns =['Appliances'])
target_data = normalised_data['Appliances']
target_data
```

```python

#splitting the data into training and test set
x_train, x_test, y_train, y_test = train_test_split(features_data, target_data, test_size= 0.3, random_state= 42)
#fitting the model to the traning set
linear_model = LinearRegression()
linear_model.fit(x_train, y_train)
#obtaining the predicted values
predicted_values = linear_model.predict(x_test)
predicted_values
```

```python
#Mean absolute error
mae= mean_absolute_error(y_test, predicted_values)
round(mae, 2)
```

```python
#Question 14
#Residual sum of squares
rss= np.sum(np.square(y_test - predicted_values))
round(rss, 2)
```

```python
#Question 15
#Root mean squared error
rmse = np.sqrt(mean_squared_error(y_test, predicted_values))
round(rmse, 3)
```

```python
#Qustion 16
#co-efficient of determination
r2_score = r2_score(y_test, predicted_values)
round(r2_score, 2)
```
